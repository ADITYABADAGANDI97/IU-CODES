```python
import sqlite3
import pandas as pd

# Connect to SQLite database
db_path = "/content/analysis.db"
conn = sqlite3.connect(db_path)

# Check existing tables in the database
query = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql(query, conn)

# Display tables
print("Tables in the database:", tables)
```

```python
query = "SELECT * FROM training_data LIMIT 5;"
df = pd.read_sql(query, conn)
df
```

```python
query = "SELECT * FROM ideal_functions LIMIT 5;"
df = pd.read_sql(query, conn)
df
```

query = "SELECT * FROM test_results LIMIT 5;"

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.palettes import Spectral4

output_notebook()

# Create a figure
p = figure(title="Test Data vs Ideal Functions", x_axis_label="X", y_axis_label="Y", width=800, height=500)

# Plot test points
p.circle(df["x"], df["y"], size=8, color="red", legend_label="Test Points")

# Plot matched ideal functions
for func_num in df["ideal_func_num"].unique():
    ideal_func_query = f"SELECT x, y{func_num} FROM ideal_functions"
    ideal_func_df = pd.read_sql(ideal_func_query, conn)
    p.line(ideal_func_df["x"], ideal_func_df[f"y{func_num}"], legend_label=f"Ideal Function {func_num}", line_width=2)

# Show plot
p.legend.click_policy = "hide"
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.palettes import Spectral4

output_notebook()

# Create a figure
p = figure(title="Test Data vs Ideal Functions",
           x_axis_label="X", y_axis_label="Y",
           width=800, height=500)

# Plot test points with updated method (using radius)
p.circle(df["x"], df["y"], radius=0.2, color="red", legend_label="Test Points")

# Plot matched ideal functions
for func_num in df["ideal_func_num"].unique():
    ideal_func_query = f"SELECT x, y{func_num} FROM ideal_functions"
    ideal_func_df = pd.read_sql(ideal_func_query, conn)
    p.line(ideal_func_df["x"], ideal_func_df[f"y{func_num}"],
           legend_label=f"Ideal Function {func_num}", line_width=2)

# Show plot
p.legend.click_policy = "hide"
show(p)
```
